**1) Очистка данных (20% оценки):**

```python
import pandas as pd

# Загрузка данных
file_path = 'Bank data bases.xlsx'
data = pd.ExcelFile(file_path)
df = data.parse('Bank DB')

# Проверка структуры данных
print("Информация о данных до очистки:")
print(df.info())
print("Описание данных до очистки:")
print(df.describe())

# Проверка и удаление дубликатов
duplicates = df.duplicated().sum()
print(f"Количество дубликатов: {duplicates}")
if duplicates > 0:
    df = df.drop_duplicates()

# Функция для поиска выбросов с использованием метода межквартильного размаха (IQR)
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
#Возвращает строки, значения которых находятся за границами.

# Функция для удаления выбросов
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
#Удаляет строки, где значения находятся за пределами границ.


# Обработка выбросов в числовых столбцах
columns_with_outliers = ['credit_amount', 'income_sum', 'month_payment', 'job_experience_year', 'Age']
for col in columns_with_outliers:
    print(f"Обработка выбросов для столбца: {col}")
    outliers = detect_outliers_iqr(df, col)
    print(f"Найдено выбросов: {len(outliers)}")
    df = remove_outliers_iqr(df, col)

# Итоговая информация о данных
print("Информация о данных после очистки:")
print(df.info())
print("Описание данных после очистки:")
print(df.describe())
```

Всего было удалено 86 строк, что составляет примерно 15.50% от исходного набора данных.

**3) Преобразование данных и проектирование признаков (20% оценки):**

```python
import pandas as pd
import numpy as np

# Загрузка данных
file_path = 'Bank data bases.xlsx'
data = pd.ExcelFile(file_path)
df = data.parse('Bank DB')

# 1. Создание новых столбцов
# Пример: Вычисление коэффициента ежемесячного платежа к доходу
df['payment_to_income_ratio'] = df['month_payment'] / df['income_sum']
#Новый столбец payment_to_income_ratio показывает, какую часть дохода клиент тратит на ежемесячный платёж.

# Преобразование категориальных данных в числовые значения
df['work_position_encoded'] = df['work_position'].astype('category').cat.codes

# 2. Использование методов фильтрации и сортировки
#Выбираются строки, где значение payment_to_income_ratio превышает 0.5.
#Эти записи представляют клиентов с высокой финансовой нагрузкой.
# Фильтрация: отбор записей с высоким коэффициентом платежа к доходу (> 0.5)
high_ratio_subset = df[df['payment_to_income_ratio'] > 0.5]

# Сортировка: сортировка по коэффициенту в порядке убывания
sorted_subset = high_ratio_subset.sort_values(by='payment_to_income_ratio', ascending=False)

# 3. Демонстрация индексации, срезов и фильтрации
# Индексация: выбор конкретных столбцов и первых 10 строк
subset_indexing = df[['statement_id', 'payment_to_income_ratio', 'work_position_encoded']].head(10)

# 4. Добавление фиктивного объединения (например, с внешними данными)
# Генерация фиктивного дополнительного DataFrame для объединения
# random_state=1: Фиксирует состояние генератора случайных чисел, чтобы выборка была воспроизводимой (одинаковые результаты при каждом запуске).
additional_data = pd.DataFrame({
    'statement_id': df['statement_id'].sample(10, random_state=1),
    'external_score': [round(x, 2) for x in np.random.uniform(0, 1, 10)]
})
# Присоединение данных к основному DataFrame
df_enriched = pd.merge(df, additional_data, on='statement_id', how='left')
#how='left': Левое объединение, добавляющее данные только для совпадающих идентификаторов.

# Итоговые действия
# Вывод первых строк каждого результата
print("Обогащенный набор данных:")
print(df_enriched.head())

print("\nОтфильтрованное подмножество данных:")
print(sorted_subset.head())

print("\nПример индексации и срезов:")
print(subset_indexing)
```

**Расширенные функции Pandas (необязательно, 10% дополнительных баллов):**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Использование пользовательских функций с apply()

# Создание пользовательской функции для классификации клиентов по финансовой нагрузке
def classify_financial_burden(ratio):
    if ratio < 0.2:
        return "Low Burden"
    elif 0.2 <= ratio <= 0.5:
        return "Moderate Burden"
    else:
        return "High Burden"

# Применение функции к коэффициенту payment_to_income_ratio
df['payment_to_income_ratio'] = df['month_payment'] / df['income_sum']
df['financial_burden_category'] = df['payment_to_income_ratio'].apply(classify_financial_burden)

# 2. Создание сводной таблицы
# Сводная таблица группирует данные по категориям финансовой нагрузки
# Сводная таблица по категориям финансовой нагрузки
pivot_table = df.pivot_table(
    index='financial_burden_category',
    values=['credit_amount', 'income_sum', 'month_payment'],
    aggfunc=['mean', 'sum']
)

# 3. Работа с временными рядами (если данные содержат временные метки)

# Создание фиктивного временного признака (в реальном случае замените на существующий)
df['loan_date'] = pd.date_range(start='2020-01-01', periods=len(df), freq='D')
#freq='D':Определяет частоту создаваемых дат.
#'D' означает, что даты будут увеличиваться с шагом в один день.


# Анализ тенденций по годам и месяцам извлекает
df['year'] = df['loan_date'].dt.year
df['month'] = df['loan_date'].dt.month

# Сумма кредитов по месяцам
monthly_trends = df.groupby(['year', 'month'])['credit_amount'].sum().reset_index()
#Метод reset_index() сбрасывает индексацию, преобразуя индекс в обычный столбец. Это удобно для работы с данными в стандартной форме DataFrame.

# Визуализация тенденций
monthly_trends['date'] = pd.to_datetime(monthly_trends[['year', 'month']].assign(day=1))
plt.figure(figsize=(10, 6))
plt.plot(monthly_trends['date'], monthly_trends['credit_amount'], marker='o')
plt.title('Тенденции суммы кредитов по месяцам')
plt.xlabel('Дата')
plt.ylabel('Сумма кредитов')
plt.grid()
plt.show()

# Вывод результатов в консоль
print("\nКатегоризация финансовой нагрузки (первые 10 строк):")
print(df[['payment_to_income_ratio', 'financial_burden_category']].head(10))

print("\nСводная таблица финансовой нагрузки:")
print(pivot_table)

print("\nТенденции суммы кредитов по месяцам:")
print(monthly_trends.head())
```
